### Machine Learning Libraries for Text Classification
The cell above imports essential libraries for a machine learning pipeline:
- **NumPy & Pandas**: Data manipulation.
- **CountVectorizer**: Converting text to numerical features.
- **MultinomialNB**: The Naive Bayes classifier.
- **Metrics**: Evaluation tools like accuracy and F1-score.

````python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,classification_report
````

# 📂 Load the Dataset

The first step is to load the training dataset into a Pandas DataFrame. This allows us to inspect and preprocess the data before building the machine learning model.

````python
df = pd.read_csv("/content/drive/MyDrive/Text Classification/train.csv")
````

# 👀 Preview the Dataset

Display the first five records to understand the structure of the dataset, available columns, and sample text entries.

````python
df.head()
````

# 📊 Dataset Information

Inspect the dataset to view the total number of records, column names, data types, and identify any missing values.

````python
df.info()
````

# 🔍 Check Missing Values

Identify null or missing values in each column to ensure data quality before preprocessing.

````python
df.isnull().sum()
````

# 🧹 Handle Missing Values

Remove rows containing missing values to create a clean dataset for model training.

````python
df.dropna(inplace=True)
````

# 🔄 Check Duplicate Records

Detect duplicate rows that may negatively affect model performance or introduce bias.

````python
df.duplicated().sum()
````

# 🗑️ Remove Duplicates

Remove duplicate records to improve data consistency and avoid redundant training samples.

````python
df.drop_duplicates(inplace=True)
````

# ✅ Verify Data Cleaning

Confirm that all missing values and duplicate records have been successfully removed from the dataset.

````python
print("Null Values",df.isnull().sum())
print("Duplicates",df.duplicated().sum())
````

# 📈 Dataset Summary Statistics

Generate descriptive statistics to gain insights into the numerical attributes of the dataset. This helps in understanding the distribution and characteristics of the data.

````python
df.describe()
````

# 🎯 Define Features and Target Variable

Separate the dataset into input features (`sentence`) and the target variable (`label`). The model will learn to predict the sentiment label based on the input text.
# ✂️ Split the Dataset

Split the dataset into training and testing sets using an 80:20 ratio. The training set is used to train the model, while the testing set is reserved for evaluating its performance on unseen data.

````python
x = df["sentence"]
y = df["label"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
````

# 🔤 Text Vectorization

Machine learning models cannot process raw text directly. CountVectorizer converts each sentence into a numerical feature vector based on the frequency of words, enabling the model to learn patterns from textual data.

````python
vectorizer = CountVectorizer()
x_train_v = vectorizer.fit_transform(x_train)
x_test_v = vectorizer.transform(x_test)
````

# 🤖 Train the Sentiment Classification Model

Train a Multinomial Naive Bayes classifier using the vectorized training data. This algorithm is widely used for text classification because it is simple, efficient, and performs well on word-frequency features.

````python
model = MultinomialNB()
model.fit(x_train_v,y_train)
````

# 🔮 Predict Sentiment Labels

Use the trained model to predict sentiment labels for the unseen test dataset.

````python
y_pred = model.predict(x_test_v)
````

# 📊 Model Evaluation

Evaluate the classifier using a classification report, which includes Precision, Recall, F1-Score, and Accuracy. These metrics provide a comprehensive assessment of the model's performance on the test data.

````python
print(classification_report(y_pred,y_test))
````

# 📝 Test the Model on Custom Input

Provide a new sentence and use the trained model to predict its sentiment. This demonstrates how the model can classify real-world text entered by a user.

````python
sample = ["Hey how are you buddy! "]
sample_v = vectorizer.transform(sample)
print(model.predict(sample_v)[0])
````

### 📌 Observation

- The dataset was successfully split into training and testing subsets.
- Text data was transformed into numerical vectors using CountVectorizer.
- A Multinomial Naive Bayes model was trained and evaluated using standard classification metrics.
- The trained model successfully predicts the sentiment of new user-provided sentences, demonstrating its practical applicability.

## 💾 Save the Trained Model

After training and evaluating the machine learning model, we save it as a `.pkl` file using the `joblib` library. This allows us to reuse the trained model later for making predictions without retraining it, which is essential for deployment in web applications such as Streamlit.

````python
import joblib

joblib.dump(model, 'emotion_model.pkl')
print("Model saved as emotion_model.pkl")
````

````python
import joblib

joblib.dump(vectorizer, 'vectorizer.pkl')
print("Vectorizer saved as vectorizer.pkl")
````
